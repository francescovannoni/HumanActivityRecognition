# Data pre-processing and Exploration Data Anaysis

```python
import numpy as np
import pandas as pd

# get the features from the file features.txt
features = list()
with open('/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/features.txt') as f:
    features = f.readlines()
print('No of Features: {}'.format(len(features)))
```

```python
# get the data from txt files to pandas dataframe
X_train = pd.read_csv("/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/train/X_train.txt", delim_whitespace=True, header=None, names=features)

# add subject column to the dataframe
X_train['subject'] = pd.read_csv('/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/train/subject_train.txt', header=None, squeeze=True)

y_train = pd.read_csv('/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/train/y_train.txt', names=['Activity'], squeeze=True)
y_train_labels = y_train.map({1: 'WALKING', 2:'WALKING_UPSTAIRS',3:'WALKING_DOWNSTAIRS',\
                       4:'SITTING', 5:'STANDING',6:'LAYING'})

# put all columns in a single dataframe
train = X_train
train['Activity'] = y_train
train['ActivityName'] = y_train_labels
train.head(10)
```

Do the same thing for test data

```python
# get the data from txt files to pandas dataffame
X_test = pd.read_csv('/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/test/X_test.txt', delim_whitespace=True, header=None, names=features)

# add subject column to the dataframe
X_test['subject'] = pd.read_csv('/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/test/subject_test.txt', header=None, squeeze=True)

# get y labels from the txt file
y_test = pd.read_csv('/home/frenk/Documenti/Artificial Intelligence/UCI_HAR_Dataset/test/y_test.txt', names=['Activity'], squeeze=True)
y_test_labels = y_test.map({1: 'WALKING', 2:'WALKING_UPSTAIRS',3:'WALKING_DOWNSTAIRS',\
                       4:'SITTING', 5:'STANDING',6:'LAYING'})


# put all columns in a single dataframe
test = X_test
test['Activity'] = y_test
test['ActivityName'] = y_test_labels
test.sample()
```

# Working on datas

```python
print('{} NaN/Null values in train'.format(train.isnull().values.sum()))
print('{} NaN/Null values in test'.format(test.isnull().values.sum()))
```

Let's check about data provided by each user


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['font.family'] = 'DejaVu Sans'

plt.figure(figsize=(20,10))
plt.title('Data from each user', fontsize=20)
sns.countplot(x='subject',hue='ActivityName', data = train)
plt.show()
```
